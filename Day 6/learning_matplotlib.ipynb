```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(1, 12)
y = np.arange(13, 24)
```

```python
plt.xlabel('years')
plt.ylabel('population of world')
plt.title("world population growth")
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 10, 0.01)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.title("trignometric chart")
plt.xlabel('trignometric values')
plt.ylabel('angle values')
plt.plot(x, y_sin, label = "sin")
plt.plot(x, y_cos, label = "cos")
plt.legend()
plt.show()
```

```python
x = np.arange(0, 10, 1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.title("trignometric chart")
plt.xlabel('trignometric values')
plt.ylabel('angle values')
plt.scatter(x, y_sin, label = "sin")
plt.scatter(x, y_cos, label = "cos")
plt.legend()
plt.show()
```
